# Data Extraction

```python
question_gen = ""
for page in data:
    question_gen += page.page_content
```

```python
question_gen
```

# Chunking

```python
from langchain.text_splitter import TokenTextSplitter
splitter_ques  = TokenTextSplitter(
    model_name= "gpt-3.5-turbo",
    chunk_size = 1000,
    chunk_overlap = 200
)
```

```python
chunk_ques_gen = splitter_ques.split_text(question_gen)
```

```python
from langchain.docstore.document import Document
doc_ques = [Document(page_content = t ) for t in chunk_ques_gen]
```

```python
prompt_extract_questions = """
You are a Python expert and your job is to extract **only the theory-based Python questions** from the following text.

### Instructions:
-
- Do not create any new questions.
- Extract only the questions that are already present in the text.
- Remove any answers or explanations.
- Each question should be on a new line.
- Do not include numbering, bullets, or extra formatting.
- Questions should be complete and clearly phrased.

### Input Text:
{text}

### Output:
List of at least 20 clean, theory-based Python questions. Each on a **separate line**.
"""
```

```python
from langchain_core .prompts import PromptTemplate
prompt_ques =  PromptTemplate(
    template = prompt_extract_questions,
    input_variables = ['text']
)
```

```python
from langchain.chains import LLMChain
question_chain = LLMChain(llm=llm, prompt=prompt_ques)
```

```python
result = question_chain.invoke({"text": question_gen})
```

```python
print(result)
```

```python
output_text = result.get("text", "")

# Split each question into a list
Ques_list = output_text.split("\n")

# Print or loop through the questions
for question in Ques_list:
    print(question)
```

# Emeddings

```python
from langchain_huggingface import HuggingFaceEmbeddings
embd = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2" )
```

# vector DB

```python
from langchain.vectorstores import FAISS
vector_store = FAISS.from_documents(doc_ques, embd)
```

# Connecting LLM to my vectore store

```python
from langchain.chains import RetrievalQA

answer_gen_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vector_store.as_retriever()  # ✅ fixed spelling
)
```

```python
for idx, question in enumerate(Ques_list, 1):
    print(f"\nQuestion {idx}: {question}")
    
    try:
        answer = answer_gen_chain.run(question)
    except Exception as e:
        answer = f"[Error fetching answer: {str(e)}]"

    print(f"Answer {idx}: {answer}")
    print("-" * 80)

    with open("interview_answers.txt", "a", encoding="utf-8") as f:
        f.write(f"Q{idx}: {question}\n")
        f.write(f"A{idx}: {answer}\n")
        f.write("=" * 80 + "\n\n")
```
